```python
import pandas as pd
```

```python
folders = ["dr0.1", "dr0.2", "dr0.3"]
sub_folders = ["pr0.4","pr0.6","pr0.8"]
total_df = pd.DataFrame()
counter = 0
for folder in folders:
    for sub_folder in sub_folders:
        file_df = pd.read_csv(folder + "\\" + sub_folder + "\\accuracies.csv")
        if counter == 0:
            total_df = file_df
            counter += 1
        else:
            total_df = pd.concat([total_df, file_df], axis=0)
```

```python
total_df
```

```python
total_df = total_df.reset_index(drop=True)
total_df = total_df.drop(['model', 'dataset', 'epochs', 'post_epochs', 'baseline_accuracy', 'high_scores_accuracy','low_scores_accuracy', 'pruned_baseline_accuracy', 'pruned_low_scores_accuracy', 'pruned_high_scores_accuracy'], axis=1)
```

```python
# pruned_low - pruned_baseline
# pruned_high - pruned_baseline
```

```python
total_df
```

```python
total_df['low-base'] = total_df['post_pruned_low_accuracy'] - total_df['post_pruned_baseline_accuracy'] 
total_df['high-base'] = total_df['post_pruned_high_accuracy'] - total_df['post_pruned_baseline_accuracy']
total_df = total_df.drop(['post_pruned_baseline_accuracy','post_pruned_low_accuracy','post_pruned_high_accuracy'], axis=1)
total_df
```

```python
total_df.to_csv('output.csv', index=False)
```
